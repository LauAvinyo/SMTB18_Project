```python
from Bio import AlignIO
```

```python
alignment = AlignIO.read(open("msa1.fasta"), "fasta")
print(alignment)
```

```python
alig= {}
for s in alignment:
    alig[s.name] = str(s.seq)

print(alig)
    
```

```python
length = len(alig['V7C8B0_PHAVU/36-160'])
```

```python
def printPos(alig, pos):
    for s in alig:
        #print(s)
        print(alig[s][pos])
```

```python
printPos(alig, 3)
```

```python
def getShannon(alig, pos):
    count = {}
    for s in alig:
        aa = alig[s][pos]
        if aa in count:
            count[aa] += 1 # i++;
        else:
            count[aa] = 1
    
    nseq = len(alig)
    for s in count:
        count[s] /= nseq
    
    h = 0
    for w in count:
        h += count[w]*log(count[w])
    
    h = -h
    return h
```

```python
getShannon(alig, 3)
```

```python
from math import log
```

```python
hs = []
for pos in range(0, length):
    hs.append(getShannon(alig, pos))
```
